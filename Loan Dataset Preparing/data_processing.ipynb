# Ordinal Encoding

## using Scikit-learn

```python
tmp_df = pd.DataFrame({"Size": ["s", "l", "m", "xl", "s", "m", "l", "l", "xl", "m", "s"]})
tmp_df.head()
```

```python
tmp_df["Size"].unique()
```

```python
ord_data = data = [["s", "m", 'l', "xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
oe.fit(tmp_df[["Size"]])
```

```python
tmp_df["Size_en"] = oe.transform(tmp_df[["Size"]])
tmp_df
```

```python
dataset = pd.read_csv("loan-train.csv")
dataset.head(3)
```

```python
dataset["Property_Area"].unique()
```

```python
dataset["Property_Area"] = dataset["Property_Area"].fillna(dataset["Property_Area"].mode()[0])
```

```python
dataset["Property_Area"].unique()
```

```python
ord_rule = [['Rural', 'Semiurban', 'Urban']]
oen = OrdinalEncoder(categories=ord_rule)
```

```python
dataset["Property_Area"] = oen.fit_transform(dataset[["Property_Area"]])
```
